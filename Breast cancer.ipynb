```python

# 1. Encode the target and drop the original
df['diagnosis_encoded'] = df['diagnosis'].map({'B': 0, 'M': 1}).astype(int)
df = df.drop(columns=['diagnosis'])

# 2. Outlier removal (Z‑score)
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop(['id','diagnosis_encoded'])
z_scores   = np.abs(stats.zscore(df[numeric_cols]))
mask       = (z_scores < 3).all(axis=1)
df_clean   = df[mask]

# Show before/after
counts_df = pd.DataFrame({
    'Stage': ['Original', 'After Outlier Removal'],
    'Count': [len(df), len(df_clean)]
})
print("Dataset Size Before/After Outlier Removal:")
display(counts_df)

# 3. Feature selection via correlation
corr_matrix = df_clean[numeric_cols].corr().abs()
upper       = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop     = [c for c in upper.columns if any(upper[c] > 0.95)]
selected_features = [c for c in numeric_cols if c not in to_drop]

print("\nDropped Features (corr>0.95):")
display(pd.DataFrame({'Feature': to_drop}))
print("\nSelected Features:")
display(pd.DataFrame({'Feature': selected_features}))

# 4. Prepare X, y (now using the numeric target!)
X = df_clean[selected_features]
y = df_clean['diagnosis_encoded']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 5. Cross‑validate
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree':        DecisionTreeClassifier(random_state=42)
}
scoring = ['accuracy','precision','recall','f1']
results = {}
for name, model in models.items():
    cv = cross_validate(model, X_train, y_train, cv=5, scoring=scoring)
    results[name] = {m: cv[f'test_{m}'].mean() for m in scoring}

results_df = pd.DataFrame(results).T.round(3)
print("\nBaseline Model (5‑fold CV) Performance:")
display(results_df)

# 6. Boxplot
fig, ax = plt.subplots()
df_clean.boxplot(column='area_mean', by='diagnosis_encoded', ax=ax)
ax.set_title('area_mean by Diagnosis (Cleaned Data)')
ax.set_xlabel('Diagnosis (0=B, 1=M)')
ax.set_ylabel('area_mean')
plt.suptitle('')
plt.tight_layout()
plt.show()
```

```python
# Hyperparameter tuning 
models_params = {
    'LogisticRegression': (
        LogisticRegression(max_iter=1000),
        {'C': [0.01, 0.1, 1, 10]}
    ),
    'DecisionTree': (
        DecisionTreeClassifier(random_state=42),
        {'max_depth': [None, 5, 10],
         'min_samples_split': [2, 5],
         'min_samples_leaf': [1, 2]}
    ),
    'RandomForest': (
        RandomForestClassifier(random_state=42),
        {'n_estimators': [100, 200],
         'max_depth': [None, 5, 10],
         'min_samples_split': [2, 5],
         'min_samples_leaf': [1, 2]}
    )
}

results = []
for name, (estimator, params) in models_params.items():
    grid = GridSearchCV(estimator, params, cv=3, scoring='roc_auc')
    grid.fit(X_train, y_train)
    best = grid.best_estimator_
    results.append({
        'Model':         name,
        'Best Params':   grid.best_params_,
        'CV ROC-AUC':    round(grid.best_score_, 3),
        'Test Accuracy': round(best.score(X_test, y_test), 3),
        'Test ROC-AUC':  round(roc_auc_score(y_test, best.predict_proba(X_test)[:,1]), 3),
    })

results_df = pd.DataFrame(results)
print("Hyperparameter Tuning & Model Comparison:")
display(results_df)
```

```python
# 1. Tuned models (unchanged)
tune_models = {
    'Logistic Regression': (
        LogisticRegression(solver='liblinear', max_iter=1000),
        {'clf__C': [0.1, 1]}
    ),
    'Decision Tree': (
        DecisionTreeClassifier(random_state=42),
        {'clf__max_depth': [5, 10]}
    ),
    'Random Forest': (
        RandomForestClassifier(random_state=42),
        {'clf__n_estimators': [50, 100], 'clf__max_depth': [5]}
    ),
}

results = []
scoring = ['accuracy', 'precision', 'recall', 'f1']

for name, (estimator, params) in tune_models.items():
    pipe = Pipeline([('clf', estimator)])
    grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    best = grid.best_estimator_
    cv_res = cross_validate(best, X_train, y_train, cv=5, scoring=scoring, n_jobs=-1)
    results.append({
        'Model':       name,
        'Best Params': grid.best_params_,
        'Accuracy':    np.round(np.mean(cv_res['test_accuracy']),  3),
        'Precision':   np.round(np.mean(cv_res['test_precision']), 3),
        'Recall':      np.round(np.mean(cv_res['test_recall']),    3),
        'F1 Score':    np.round(np.mean(cv_res['test_f1']),        3),
    })

# 2. Default ensemble model: only XGBoost (wrapped in Pipeline)
default_model = Pipeline([
    ('clf', XGBClassifier(
        use_label_encoder=False,
        eval_metric='logloss',
        random_state=42
    ))
])
cv_results = cross_validate(default_model, X_train, y_train,
                            cv=5, scoring=scoring, n_jobs=-1)
results.append({
    'Model':       'XGBoost (default)',
    'Best Params': 'default',
    'Accuracy':    np.round(np.mean(cv_results['test_accuracy']),  3),
    'Precision':   np.round(np.mean(cv_results['test_precision']), 3),
    'Recall':      np.round(np.mean(cv_results['test_recall']),    3),
    'F1 Score':    np.round(np.mean(cv_results['test_f1']),        3),
})

# Display and plot
res_df = pd.DataFrame(results)
print("Tuned & Ensemble Model Performance (No SVM):")
display(res_df)

ax = res_df.set_index('Model')['Accuracy'].plot.bar()
ax.set_ylabel('Accuracy')
ax.set_title('Model Accuracy Comparison (No SVM)')
plt.tight_layout()
plt.show()
```
